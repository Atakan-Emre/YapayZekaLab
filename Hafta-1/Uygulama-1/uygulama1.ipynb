## 5) Aşama (d)
**(3,4) boyutunda bir dizi oluşturup bu dizinin boyutunu (6,2) olacak şekilde değiştiriniz.**

```python
# 3x4 boyutunda bir NumPy dizisi oluşturalım
arr = np.arange(12).reshape(3,4)
print("3x4 dizi:\n", arr, "\n")

# Boyutu 6x2 olacak şekilde yeniden şekillendirelim
arr_reshaped = arr.reshape(6,2)
print("6x2 dizi:\n", arr_reshaped)
```

## 6) Aşama (e)
**İki tane (3,3) boyutunda rastgele sayılardan oluşan dizi oluşturun. Bu dizileri yatay ve dikey olacak şekilde istifleyin (stack).**

```python
# İki adet 3x3 rastgele dizi oluşturalım (0-50 arasında örnek)
arr1 = np.random.randint(0, 50, (3,3))
arr2 = np.random.randint(0, 50, (3,3))

print("arr1:\n", arr1)
print("\narr2:\n", arr2)

# Yatay istif (hstack)
h_stacked = np.hstack((arr1, arr2))

# Dikey istif (vstack)
v_stacked = np.vstack((arr1, arr2))

print("\nYatay istif (hstack):\n", h_stacked)
print("\nDikey istif (vstack):\n", v_stacked)
```
