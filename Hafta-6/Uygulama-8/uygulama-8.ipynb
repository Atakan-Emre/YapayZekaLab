```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import confusion_matrix, classification_report

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Dense, Conv1D, Dropout, MaxPooling1D,
                                     Flatten, BatchNormalization, Input)
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import SGD, RMSprop, Adam

# -----------------------------------------------------
# 1) LEAF VERİ SETİ (train1.csv, test1.csv)
# -----------------------------------------------------

# 1A) Veri Yükleme, Hazırlık
train_leaf_df = pd.read_csv("train1.csv")
test_leaf_df = pd.read_csv("test1.csv")

le = LabelEncoder().fit(train_leaf_df["species"])
labels_leaf = le.transform(train_leaf_df["species"])  # 0,1,2,... sınıf indeksleri
classes_leaf = list(le.classes_)  # Sınıf isimleri (türler)

train_leaf_df = train_leaf_df.drop(["id", "species"], axis=1)
test_leaf_df = test_leaf_df.drop(["id"], axis=1)

nb_features_leaf = train_leaf_df.shape[1]
nb_classes_leaf = len(classes_leaf)

# Ölçeklendirme
scaler_leaf = StandardScaler().fit(train_leaf_df)
train_leaf = scaler_leaf.transform(train_leaf_df)
test_leaf = scaler_leaf.transform(test_leaf_df)

# 1B) Kendi Modelinizi Oluşturma (Örnek: Conv1D tabanlı)
def create_leaf_model():
    model = Sequential()
    model.add(Input(shape=(nb_features_leaf, 1)))
    model.add(Conv1D(256, 3, activation="relu"))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(2))
    model.add(Conv1D(128, 3, activation="relu"))
    model.add(MaxPooling1D(2))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(256, activation="relu"))
    model.add(Dropout(0.5))
    model.add(Dense(nb_classes_leaf, activation="softmax"))
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

# 2) Çapraz Doğrulama (5 Katlı K-Fold)
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = []

for fold_index, (train_idx, val_idx) in enumerate(kfold.split(train_leaf, labels_leaf)):
    print(f"\n=== K-Fold {fold_index+1} ===")
    X_train_fold, X_val_fold = train_leaf[train_idx], train_leaf[val_idx]
    y_train_fold, y_val_fold = labels_leaf[train_idx], labels_leaf[val_idx]
    
    # One-hot
    y_train_fold_oh = to_categorical(y_train_fold, nb_classes_leaf)
    y_val_fold_oh = to_categorical(y_val_fold, nb_classes_leaf)

    # Modeli sıfırdan oluştur
    fold_model = create_leaf_model()

    # Eğit
    fold_model.fit(X_train_fold.reshape(-1, nb_features_leaf, 1),
                   y_train_fold_oh,
                   epochs=5,  # Örnek olarak kısaltılmış
                   validation_data=(X_val_fold.reshape(-1, nb_features_leaf, 1), y_val_fold_oh),
                   verbose=1)
    # Değerlendir
    scores = fold_model.evaluate(X_val_fold.reshape(-1, nb_features_leaf, 1),
                                 y_val_fold_oh, verbose=0)
    print(f"Fold {fold_index+1} Accuracy: {scores[1]*100:.2f}%")
    cv_scores.append(scores[1])

print("\n----- K-Fold Sonuçları -----")
print("Fold Accuracies:", [f"{s*100:.2f}%" for s in cv_scores])
print("Ortalama Accuracy: %.2f%%" % (np.mean(cv_scores)*100))

# 3) Final Train / Valid Split + F1, Precision, Recall, Specificity

# Train/Valid ayırma
X_train_leaf, X_valid_leaf, y_train_leaf_raw, y_valid_leaf_raw = train_test_split(
    train_leaf, labels_leaf, test_size=0.2, random_state=42
)
y_train_leaf_oh = to_categorical(y_train_leaf_raw, nb_classes_leaf)
y_valid_leaf_oh = to_categorical(y_valid_leaf_raw, nb_classes_leaf)

# Modeli tekrar oluşturup eğit
final_model_leaf = create_leaf_model()
history_leaf = final_model_leaf.fit(
    X_train_leaf.reshape(-1, nb_features_leaf, 1),
    y_train_leaf_oh,
    epochs=5,  # kısaltılmış
    validation_data=(X_valid_leaf.reshape(-1, nb_features_leaf, 1), y_valid_leaf_oh),
    verbose=1
)

# Tahmin
y_pred_leaf_oh = final_model_leaf.predict(X_valid_leaf.reshape(-1, nb_features_leaf, 1))
y_pred_leaf = np.argmax(y_pred_leaf_oh, axis=1)   # Sınıf indeksleri
y_true_leaf = y_valid_leaf_raw

print("\n----- Classification Report (Leaf) -----")

# Confusion Matrix
print(classification_report(
    y_true_leaf, 
    y_pred_leaf,
    labels=range(len(classes_leaf)),   # 0'dan 98'e kadar tüm sınıfları kullan
    target_names=classes_leaf          # 99 sınıf ismi
))

conf_mat_leaf = confusion_matrix(
    y_true_leaf,
    y_pred_leaf,
    labels=range(len(classes_leaf))    # Confusion matrix de tüm sınıfları baz alır
)
print("Confusion Matrix:\n", conf_mat_leaf)

# Ek metrikler (Sensitivity, Specificity)
# Sensitivity(Recall) = TP/(TP+FN) => classification_report'ta recall.
# Specificity = TN/(TN+FP). Bunu her sınıf için hesaplamak isterseniz, döngü veya multi-label approach gerekir.
# Örnek olarak 'binary' metrik gösterimi lazımsa, Leaf veri seti çok sınıflı olduğu için her sınıf ayrı hesaplanır.
# (İsteğe bağlı: Aşağıda 1 sınıfa karşı "rest" gibi binary yaklaşımla specificity gösterilebilir, ama multi-class'ta
#  net bir tanım yapmak istiyorsanız macro vs. approach gerekir.)

# -----------------------------------------------------
# 4) DİYABET VERİ SETİ (diabetes.csv)
# -----------------------------------------------------
diabetes_df = pd.read_csv("diabetes.csv")
X_diab = diabetes_df.drop("Outcome", axis=1)
y_diab = diabetes_df["Outcome"]

# Ölçeklendirme
scaler_diab = StandardScaler().fit(X_diab)
X_diab = scaler_diab.transform(X_diab)

# Train/Valid bölme
X_train_diab, X_valid_diab, y_train_diab_raw, y_valid_diab_raw = train_test_split(
    X_diab, y_diab, test_size=0.2, random_state=42
)

# 1DESA modeliniz (Basit Dense Katman Örneği):
model_diabetes = Sequential()
model_diabetes.add(Dense(64, activation='relu', input_shape=(X_train_diab.shape[1],)))
model_diabetes.add(Dropout(0.2))
model_diabetes.add(Dense(1, activation='sigmoid'))  # Tek çıkış nöronu (binary classification)

model_diabetes.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

history_diabetes = model_diabetes.fit(
    X_train_diab, y_train_diab_raw,
    epochs=10,  # Kısaltılmış
    validation_data=(X_valid_diab, y_valid_diab_raw),
    verbose=1
)

# Doğrulama skoru
loss_diab, acc_diab = model_diabetes.evaluate(X_valid_diab, y_valid_diab_raw, verbose=0)
print(f"\nDiyabet Modeli - Validation Accuracy: {acc_diab*100:.2f}%, Loss: {loss_diab:.4f}")

# F1-Skoru (binary)
y_pred_diab_prob = model_diabetes.predict(X_valid_diab)
y_pred_diab_bin = (y_pred_diab_prob >= 0.5).astype(int)

# classification_report içinde precision, recall, f1 bulmak
print("\n----- Diyabet Classification Report -----")
print(classification_report(y_valid_diab_raw, y_pred_diab_bin))

# Öğrenme Eğrileri
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history_diabetes.history['loss'], label="Eğitim Kaybı")
plt.plot(history_diabetes.history['val_loss'], label="Doğrulama Kaybı")
plt.title("Loss - Diyabet Modeli")
plt.legend()

plt.subplot(1,2,2)
plt.plot(history_diabetes.history['accuracy'], label="Eğitim Doğruluğu")
plt.plot(history_diabetes.history['val_accuracy'], label="Doğrulama Doğruluğu")
plt.title("Accuracy - Diyabet Modeli")
plt.legend()
plt.show()

# -----------------------------------------------------
# 5) Farklı optimizasyon algoritmalarıyla deney
# -----------------------------------------------------
optimizers = [SGD(), RMSprop(), Adam()]
optimizer_names = ["SGD", "RMSprop", "Adam"]
results = {}

for opt, opt_name in zip(optimizers, optimizer_names):
    # Yeni model oluştur (binary classification)
    temp_model = Sequential()
    temp_model.add(Dense(64, activation='relu', input_shape=(X_train_diab.shape[1],)))
    temp_model.add(Dropout(0.2))
    temp_model.add(Dense(1, activation='sigmoid'))

    temp_model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy"])
    temp_history = temp_model.fit(
        X_train_diab, y_train_diab_raw,
        epochs=5,  # Kısaltılmış
        validation_data=(X_valid_diab, y_valid_diab_raw),
        verbose=0
    )
    val_acc = np.mean(temp_history.history['val_accuracy'])  # son epoch'ların ortalaması
    results[opt_name] = val_acc

best_opt = max(results, key=results.get)
print("\n----- Farklı Optimizasyon Karşılaştırması -----")
for k,v in results.items():
    print(f"{k}: {v*100:.2f}% doğruluk (ortalama val_acc)")
print(f"En iyi algoritma: {best_opt} -> {results[best_opt]*100:.2f}%")
```
